#### Treating Null Values

```python
#we will remove null values more than 50% according to Industry standards


#assigning the null value percentages of each column to a variable nul_p
nul_p=((app_df.isnull().sum()*100)/305711)

#columns with more than 50% null value content. Assigning a variable col_n to store the column names
col_n=list(app_df.columns[nul_p>50])

#removing columns with more than 50% null value content
app_df.drop(labels=col_n, axis=1, inplace=True)
app_df.shape
```

```python
'A total of 122-78=44 columns were removed'
```

```python
#updating nul_p variable 
nul_p=((app_df.isnull().sum()*100)/305711)

```

```python
#Checking columns with less than or equal to 13% of null value contents & assigning it to a new variable nul
nul=app_df.columns[nul_p<=13]
nul
```

#### Checking for errors in datatype & correcting them

```python
#Before proceeding, lets take a look at the app_df to bring in the focus
app_df.describe()
```

```python
# We can see some -ve values in the days related column by using the describe funcion. SO lets rectify them

#lets specifically look at the columns that we need to investigate further & work on them
d=[column for column in app_df if column.startswith('DAYS')]
d
```

```python
#Checking each column 
print(app_df.DAYS_BIRTH.unique())
print(app_df.DAYS_EMPLOYED.unique())
print(app_df.DAYS_REGISTRATION.unique())
print(app_df.DAYS_ID_PUBLISH.unique())
print(app_df.DAYS_LAST_PHONE_CHANGE.unique())
```

```python
#Each column has -ve values,so by using the abs function we can resolve the issue
 
app_df.DAYS_BIRTH=abs(app_df.DAYS_BIRTH)
app_df.DAYS_EMPLOYED=abs(app_df.DAYS_EMPLOYED)
app_df.DAYS_REGISTRATION=abs(app_df.DAYS_REGISTRATION)
app_df.DAYS_ID_PUBLISH=abs(app_df.DAYS_ID_PUBLISH)
app_df.DAYS_LAST_PHONE_CHANGE=abs(app_df.DAYS_LAST_PHONE_CHANGE)
```

#### Binning Continuous Variables

```python
# Binning AMT_INCOME_TOTAL & storing it into a new column INCOME_CATEGORY
app_df['INCOME_CATEGORY'] = pd.qcut(app_df.AMT_INCOME_TOTAL, q=[0, 0.2, 0.4, 0.6, 0.8, 1], labels=['VERY_LOW', 'LOW', "MEDIUM", 'HIGH', 'VERY_HIGH'])
app_df['INCOME_CATEGORY'].head(10)
```

```python
# Binning DAYS_BIRTH & storing it into a new column AGE_GROUP

#We earlier saw that the age was in days. So we need to convert it to years
app_df.DAYS_BIRTH=app_df.DAYS_BIRTH.apply(lambda x:int(x/365))
```

```python
# Binning DAYS_BIRTH & storing it into a new column AGE_GROUP
app_df['AGE_GROUP']=pd.cut(app_df.DAYS_BIRTH, bins=[18,30,45,60,100], labels=['Youth','Young_Adult', 'Adult', 'Senior_Citizen'])
app_df['AGE_GROUP'].value_counts()
```
